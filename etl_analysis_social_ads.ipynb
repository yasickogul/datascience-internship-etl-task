```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://raw.githubusercontent.com/yasickogul/datascience-internship-etl-task/main/social_ads.csv"
```

```python
#Extract
df = pd.read_csv(url)
df.head()
```

```python
print("Rows and Columns:", df.shape)
```

TASK 1

Columns

Age: Age of the individual (numeric)

EstimatedSalary: Annual income (numeric)

Purchased: Whether the person purchased the product (1 = Yes, 0 = No)

Number of columns = 3
Number of rows = 400

```python
#TASK 2
#Transform
# Check for nulls
print(df.isnull().sum())

# Check duplicates
df.drop_duplicates(inplace=True)

#Load
df.to_csv('cleaned_social_network_ads.csv', index=False)
```

```python
df.head()
```

```python
# Print shape
print("Dataset Shape:", df.shape)

# Print descriptive stats rounded to 2 decimals
print("\nDescriptive Statistics:")
print(df.describe().round(2))

# Purchase distribution
print("\nPurchase Distribution:")
print(df['Purchased'].value_counts())
```

```python
sns.countplot(data=df, x="Purchased")
plt.title("Purchase Distribution")
plt.xlabel("Purchased (0 = No, 1 = Yes)")
plt.ylabel("Number of Users")
plt.show()

```

```python
# Age vs Purchase
sns.boxplot(data=df, x="Purchased", y="Age")
plt.title("Age Distribution by Purchase Status")
plt.show()
```

```python
#Estimated Salary vs Purchase
sns.boxplot(data=df, x="Purchased", y="EstimatedSalary")
plt.title("Salary Distribution by Purchase Status")
plt.show()
```
